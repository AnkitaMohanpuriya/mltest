If we are given two matrices of size (n,k) and (k,m) and are multiplied, we get a (n,m) sized matrix which will contain n*m items

Then,
Z[i,j]=X[i,1]*Y[1,j]+X[i,2]*Y[2,j]+X[i,3]*Y[3,j].....+X[i,k]*[k,j]
and such operation is to be done for finding all the elements in the matrix, i.e, n* m elements
and in the given operation, k multiplications are done and (k-1) additions are done
therefore to find all the elements
number of multiplications will be k* n* m
and number of additions will be (k-1)* n* m

```python
#Question 11.

import numpy as np
import pandas as pd


high = 2
low = 1
m,n = 20,5
a = (high - low)*np.random.rand(m,n) + low
df=pd.DataFrame(a)

df.columns = ["a","b","c","d","e"]
df['f'] =  df[['a', 'b','c','d','e']].sum(axis=1)
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

print(df)
    
num_lt8 = df[df['g'] == 'LT8'].shape[0]
print("Number of rows 'LT8' in column 'g':", num_lt8)

std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print("Deviation ('LT8'):", std_lt8)
print("Deviation ('GT8'):", std_gt8)


```
